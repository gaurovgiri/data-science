```python
import matplotlib as mlp
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
plt.show() # shows the diagram with figure and axes
```

```python
x = np.linspace(0,10,1000) # creates a evenly spaced array of 1000 numbers from 0 to 10
y = np.sin(x)
fig = plt.figure()
ax = plt.axes()
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('function sin(x)')
plt.show()
```

```python

# x = np.linespace(0,10,1000)
plt.plot(x, np.sin(x), color='k')
plt.plot(x, np.cos(x), color='r',linestyle='--')
plt.xlim([0,1.5])
plt.ylim([0,2])
plt.show()
```

```python
#x = np.linespace(0,10,1000)
plt.plot(x,np.sin(x),'k:',label='sin(x)')
plt.plot(x,np.cos(x),'r--',label='cos(x)')
plt.legend() #will create legend
plt.show()
```

```python
plt.scatter(x,np.sin(x))
plt.show()
```

```python
age = np.array((16,17,18,19,20,21,15,22))
height = np.array([152,162,155,157,158,159,173,169])
plt.scatter(age,height,color='b', marker='<')
plt.xlabel('age')
plt.ylabel('height')
plt.title('Students')
plt.show()
```
